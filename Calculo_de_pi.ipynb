##Calculo de π por simulación.
Para obtener las **áreas** por simulación hay que imaginarnos **gotas de lluvia chocando contra nuestras figuras**, la proporción de gotas en el circulo entre las del cuadrado multiplicado por 4 nos darán una aproximación a pi.






```python
#Hecho con libreria seaborn y matplot
from random import random as rd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
#Numero de gotas
N=1000000
#gotas iniciales dentro del circulo 
n=0
lx=[]
ly=[]

#Lanzamiento de las gotas
for i in range(1,N+1):
  #Asignandole un valor aleatorio entre 0 y 1 a "x" y "y" 
  x=rd()
  y=rd()
  #Si caen dentro del circulo agregar 1 a gotas iniciales
  if x**2+y**2<1:
     n=1+n
  if i%1000==0:
    lx.append(i)
    ly.append(4*n/i)
    
sns.lineplot(lx,ly)
plt.ylabel("proporcion de gotas")
plt.xlabel("Numero de gotas")
plt.show()
print(4*n/N)
```
